```python
import os
import openai
openai.api_key = "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"
# openai.api_key = os.getenv("OPENAI_API_KEY")

def get_response(in_prompt):
    response = openai.Completion.create(
    model="text-davinci-003",
    prompt="In english," + in_prompt,
    temperature=0.7,
    max_tokens=4000-len(in_prompt)-10,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0
    )

    return(response['choices'][0]['text'])
```

```python
def countWords(s):
 
    # Check if the string is null
    # or empty then return zero
    if s.strip() == "":
        return 0
 
    # Splitting the string
 
    words = s.split()
 
    return len(words)
```

```python
Response_dir="D:\\Blender NLP Intern\\Data\Response Summarization"
count=0
for fname in os.listdir(Response_dir):
    with open(Response_dir+"\\"+fname, 'r',encoding="utf8") as f:
        data = json.load(f)
        print(fname," file done")
        cw=countWords(data['prompt'])
        if(cw):
            try: 
                response_summary=get_response(data['prompt'])
                data['response_summary']=response_summary
            except:
                print(fname," not done")
                data['response_summary']="Response couldn't generated because prompt was too long( "+str(cw)+"words )"
        json_file = json.dumps(data)

        with open(Response_dir+"\\"+fname, 'w',encoding="utf8") as fw:
            fw.write(json_file)
```

```python
def print_nicely(text):
    words=text.split()
    count=0
    line=""
    for word in words:
        line = line + " "+word
        count = count + 1
        if count>10:
            line = line + " \n"
            count=0
    return line
```
